```python
import pandas as pd
import csv

f = open('location.csv','r')
g = open('SampleLocation.csv','r') #Query locations

workable = list(csv.reader(f))
workable_2 = list(csv.reader(g))

for element in workable: #We need to pad f with a blank column in order to fill this last column with serial numbers
    element.append('')
    

Relevant = pd.DataFrame(workable)
print(Relevant)
    
Not_found = []
for element_2 in workable_2: #Now, we simply match panels locations
    string2 = element_2[0] #Broken panel location
    sn = element_2[1] #Broken panel serial number
    a=0 #Used to break the for loop if the panel is already found
    b=0 #Checks to make sure string2 was mapped
    counter2 = 0 # Counter used to track the list in the layout
    for element in workable:#Looping through all the lists in the layour
        counter = 0 #counter used to track the element in the list in the layout
        if a == 1:
            break
        for i in element:
            if i == string2: #If we find a match, we fill in the serial number
                element[counter] = 'X' + string2 #Change for Excel highlighting purposes
                element[counter+1] = sn #Filled in serial number
                a = 1
                b = 1
                break
            counter = counter + 1 #increment the counter
        counter2 = counter2 +1 #increment the counter
    if b==0:
        Not_found.append(string2)
        
my_df = pd.DataFrame(workable)

my_df.to_excel('my_csv.xlsx', index=False, header=False) #print to a csv
if Not_found != 0:
    print('The following locations were not mapped:')
    print(Not_found)
    print('Check the formatting')

```
